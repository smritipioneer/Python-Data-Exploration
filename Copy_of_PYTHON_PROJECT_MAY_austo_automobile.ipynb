```python
# Filter salaried individuals
salaried_df = df[df['Profession'].str.contains('Salaried', case=False, na=False)]

# Calculate probability: P(buying Sedan | salaried)
# Assuming 'Make' indicates car type, and 'Personal_loan' indicates purchase
sedan_buyers = salaried_df[salaried_df['Make'].str.contains('Sedan', case=False, na=False)]

prob_sedan_given_salaried = len(sedan_buyers) / len(salaried_df) if len(salaried_df) > 0 else 0
print(f"Likelihood of a salaried person buying a Sedan: {prob_sedan_given_salaried:.2%}")
```

####** Analysis of question 3. What evidence or data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = r"/content/drive/MyDrive/Colab Notebooks/python project great learning month1/austo_automobile .csv"
data = pd.read_csv(file_path)

# Inspect the first few rows of the dataset
#print(data.head())

# Filter data for salaried males
salaried_males = data[(data['Gender'] == 'Male') & (data['Profession'] == 'Salaried')]

# Count SUV and Sedan sales for salaried males
suv_sales = salaried_males[salaried_males['Make'] == 'SUV'].shape[0]
sedan_sales = salaried_males[salaried_males['Make'] == 'Sedan'].shape[0]

# Print the results
print(f"SUV sales to salaried males: {suv_sales}")
print(f"Sedan sales to salaried males: {sedan_sales}")

# Create a bar chart
vehicle_types = ['SUV', 'Sedan']
sales_counts = [suv_sales, sedan_sales]

plt.bar(vehicle_types, sales_counts, color=['blue', 'orange'])
plt.title('Sales Comparison for Salaried Males')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Sales')
plt.show()

# Compare and conclude
if suv_sales > sedan_sales:
    print("Evidence supports that a salaried male is an easier target for an SUV sale over a Sedan sale.")
else:
    print("Evidence does not support that a salaried male is an easier target for an SUV sale over a Sedan sale.")
```

# ** Observation on Customer Demographics and Purchasing Behavior**

**Overview:**

This report provides insights into the demographic and purchasing patterns of our customers based on recent observations and data analysis.

**1. Marital Status and Purchasing Behavior:**

- Married customers tend to purchase more products from the company.
- Married individuals exhibit higher purchasing frequency compared to unmarried customers.

**2. Employment and Income Level:**

- Salaried customers are purchasing more from the company, indicating a preference or higher engagement among employed individuals.
- The total salary distribution among customers is right-skewed, suggesting that while most customers earn within a certain range, there are customers with significantly higher salaries.

**3. Price Distribution of Vehicles:**

- The distribution of car prices is right-skewed. This indicates that most cars purchased are within an average price range, but there are some purchases of higher-cost vehicles.
- The higher-end car purchases contribute to the right skewness in the price distribution.

**4. Customer Education Level:**

- A greater number of customers are postgraduate degree holders compared to graduates.
- This trend may reflect the educational background of our customer base and could influence purchasing preferences.

**5. Age Distribution:**

- The age distribution of customers is right-skewed, indicating that the majority of our customers are relatively young.
- The number of customers decreases as age increases, suggesting a younger customer demographic.

**Conclusion:**

The data indicates that our primary customer base is young, educated, and employed individuals, with a significant proportion being married. These customers tend to have higher incomes and are more active in purchasing from the company, especially in higher-priced segments. Understanding these patterns can help tailor marketing strategies and product offerings to better serve our key customer segments.

**Recommendations:**

- Focus marketing efforts on young, postgraduate, married, and salaried individuals.
- Consider expanding premium vehicle options to cater to the higher salary and higher-end car buyers.
- Develop targeted campaigns for older customers to increase engagement across all age groups.
